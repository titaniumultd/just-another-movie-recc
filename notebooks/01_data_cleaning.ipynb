```python
import pandas as pd

# Load the dataset
movies = pd.read_csv('../data/ml-latest-small/movies.csv')
ratings = pd.read_csv('../data/ml-latest-small/ratings.csv')

# Print the first few rows to check if the data is loaded correctly
print("Movies Data:")
print(movies.head())
print("\nRatings Data:")
print(ratings.head())

# Check for missing data
print("\nMovies Data Missing Values:")
print(movies.isnull().sum())
print("\nRatings Data Missing Values:")
print(ratings.isnull().sum())

# Remove duplicates
movies.drop_duplicates(inplace=True)
ratings.drop_duplicates(inplace=True)

# Check data types
print("\nMovies Data Types:")
print(movies.dtypes)
print("\nRatings Data Types:")
print(ratings.dtypes)

# Save cleaned dataframes to new csv files
movies.to_csv('../data/cleaned_movies.csv', index=False)
ratings.to_csv('../data/cleaned_ratings.csv', index=False)
```
